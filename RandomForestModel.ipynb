```python
import numpy as np
import pandas as pd

Train_data = pd.read_csv('train.csv')

Test_data = pd.read_csv('test.csv')

Train_data.head()
```

```python
Y_train = Train_data.Survived

X_train = pd.get_dummies(Train_data.drop(['Survived','Name','Ticket','Fare','Cabin','Embarked','Age'],axis = 1))

X_test = pd.get_dummies(Test_data.drop(['Name','Ticket','Fare','Cabin','Embarked','Age'],axis =1))

X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators = 100 , max_depth = 5,random_state = 1)

model.fit(X_train,Y_train)

preds = model.predict(X_test)

output = pd.DataFrame({'PassangerId':X_test.PassengerId, 'Survived': preds})

output.to_csv('submission.csv',index = False)

print(output)
```

```python
#becouse we dont have any survival in test.csv 

from sklearn.model_selection import train_test_split

X_train_new ,X_test_new ,Y_train_new ,Y_test_new = train_test_split(X_train,Y_train,random_state =1)
```

```python
#RandomForestRegressor give us continuous output
#RandomForestClassifier give us binary output (0-1)

from sklearn.ensemble import RandomForestClassifier

new_model = RandomForestClassifier(n_estimators =100,max_depth =5,random_state =1)

new_model.fit(X_train_new, Y_train_new)
#converting those into binary classification 
new_preds = new_model.predict(X_test_new)
```

```python
from sklearn.metrics import accuracy_score 

accuracy = accuracy_score(Y_test_new,new_preds)
```
