```python
accuracies = {
    'Logistic Regression': 0.7847,
    'K-Nearest Neighbors': 0.5874,
    'Decision Tree': 0.7623,
    'Random Forest': 0.7713
}
```

```python
import matplotlib.pyplot as plt

# Model names and corresponding accuracies
models = list(accuracies.keys())
scores = list(accuracies.values())

plt.figure(figsize=(8, 4))
bars = plt.bar(models, scores, color=['red', 'blue', 'green', 'purple'])

# Add accuracy values above each bar
for bar, acc in zip(bars, scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, f'{acc:.4f}', ha='center', fontsize=12, color='black')

plt.ylim(0, 1)
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
